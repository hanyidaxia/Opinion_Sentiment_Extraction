```python
def get_shoes_similarit(shoes_1, shoes_2, df):
    """
    @does: will return similarity value between each attribute of each pairs of shoes
    """
    from scipy import spatial
    shoes_1_x = df.loc[shoes_1]
    shoes_2_x = df.loc[shoes_2]
    similarities = []
    for att1, att2 in zip(shoes_1_x, shoes_2_x):
        if (0.001 in att1) and (0.001) in att2:
            similarities.append(0)
        else:
            similarities.append(1 - spatial.distance.cosine(att1, att2))
    return similarities
```
